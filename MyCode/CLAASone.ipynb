```python
Write a Python program that asks the user for a number and prints the
sum of all numbers from 1 to the input number using a loop. Include error
handling to ensure the input is a positive integer.
```

```python
import numpy as np
sum_total=0;
x=int(input("Enter a positive number:"))
if (x<0):
    print("Sorry!x cannot be negative")
else:
    for i in range(1,x):
        sum_total+=i
    print(sum_total)
```

Using the Numpy package, generate a 10 × 10 array of random integers
between 1 and 100. Write a function that replaces all elements in the array
that are greater than 50 with the value −1. Demonstrate this functionality
with an example.

```python
import numpy as np
arr=np.random.randint(1,100,(10,10))
print(arr.shape)
for i in range(arr.shape[0]):
  for j in range(arr.shape[1]):
    if arr[i,j]>50:
       arr[i,j]= -1
print(arr)
```

```python
import matplotlib.pyplot as plt
x = np.linspace(0, 2 * np.pi, 100)
x.shape
y = np.cos(x)
z= np.sin(x)
plt.plot(z,y)
print(z)
```
